```
from math import factorial
import itertools
import numpy as np
```

```
def erat2( ):
    D = {  }
    yield 2
    for q in itertools.islice(itertools.count(3), 0, None, 2):
        p = D.pop(q, None)
        if p is None:
            D[q*q] = q
            yield q
        else:
            x = p + q
            while x in D or not (x&1):
                x += p
            D[x] = p

def get_primes_erat(n):
    return list(itertools.takewhile(lambda p: p<n, erat2()))

def wilsonPrimeRemainder(x):
    return (factorial(x-1)+1) % (x**2)
```

```
df = np.array([[0, 0]])
```

```
def getRem(limit, df):
    primes = get_primes_erat(limit)
    for p in primes:
        rem = wilsonPrimeRemainder(p)
        tmp = np.array([[p, rem]])

        if rem == 0 and p > 563:
            np.savetxt('wp.csv', tmp, delimiter=',')
            files.download('wp.csv')

        print(p, rem)

        df = np.concatenate((df, [[p, rem]]))
    return df
```

```
df = getRem(10_000_000, df)

np.savetxt('remainders.csv', df, delimiter=',')
files.download('remainders.csv')
```
